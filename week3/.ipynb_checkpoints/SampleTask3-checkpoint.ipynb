```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('text.txt', delimiter = "\t", names=['s','t']) # only 20 lines from the SMScollectiontext
df
```

```python
list(df)
```

```python
import re
df['l'] = df['t'].str.len()
df['f'] = df.t.str.extract('([A-Z_]+)')
df['fl'] = df['f'].str.len()
df2 = df.t.str.extractall('([A-Z_]+)')  # extracting uppercase alphabets
```

```python
df2.index.names = ['text','no']
df2
```

```python
df['uc'] = df2.groupby(level=[0]).sum()
df['ucl'] = df['uc'].str.len()
```

```python
list(df)
```

```python
df['ucr'] = df['ucl'].div(df['l'])
df[['ucr', 'ucl', 'l']]
```

```python
import re
df['l'] = df['t'].str.len()
df['numf'] = df.t.str.extract('([0-9]+)')
df['numfl'] = df['f'].str.len()
df3 = df.t.str.extractall('([0-9]+)')  # extracting digits from the line
```

```python
df3.index.names = ['text','no']
df3
```

```python
df['ucnum'] = df3.groupby(level=[0]).sum()
df['ucnum']
```

```python
df['ucnuml'] = df['ucnum'].str.len()
df['ucnuml']
```

```python
list(df)
```

```python
df.drop(df.index[(df.ucnuml>8)],axis=0,inplace=True) # removeing row with more than 8 uppercase letter
```

```python
df
```

```python
df.drop(df.index[(df.ucl>5)],axis=0,inplace=True) # removeing row with more than 8 uppercase letter
```
