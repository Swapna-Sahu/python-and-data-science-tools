```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('responses.csv', delimiter = ",") # read in the data and print it
df # Note that the first line of the csv-file is interpreted as variable names (which fits in this case!). 
```

```python
df.dtypes # get data types of each column of the data frame (object is string)
```

```python
df['min'] = pd.to_numeric(df['min'], errors = 'coerce') # change min from string to numeric
df
```

```python
df = df.loc[df['exp'].isin(['Yes','No'])]  # remove the row having both "yes"and "no"
df
```

```python
df['time'] = df['min']*60 + df.sec # make a time variable
df
```

```python
df3=df.loc[df.type == 'Greek']
df4=df.loc[df.type == 'Latin']
```

```python
df3
```

```python
df4
```

```python
greek = df3['time'].mean(axis = 0, skipna = True) # mean of Greek
greek
```

```python
latin = df4['time'].mean(axis = 0, skipna = True) # mean of Latin
latin
```

```python
if greek < latin:
    print('Greek were faster')
else:
    print('Latin were faster')
```
